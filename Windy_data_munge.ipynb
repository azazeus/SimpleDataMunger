## Another simple data munger for wind conditions from https://www.windy.com

### Uses selenium and chromedriver to open a webpage and read off fluorescence absorption and emission values for all fluorophores and dyes available on that page.
### The read values are then written to a Excel file for using in further calculations for overlap and filter transmission calculations.

```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
driver=webdriver.Chrome("C:\\Users\\Amarb\\Downloads\\chromedriver_win32\\chromedriver.exe")
```

```python
n_steps=5
wind_data=pd.DataFrame()
for dsteps in range(0,n_steps,1):
    lat = 18.08467-dsteps*(18.08467 - 10)/n_steps;
    longit = -118-dsteps*(-118.08467 + 130)/n_steps;
    driver.get("https://www.windy.com/"+str(lat)+"/"+str(longit));
    sleep(1.0)
    day_list=[];
    time_list=[];
    wind_list=[];
    gust_list=[];
    lat_list = [];
    long_list=[];
    wdir_list = [];
    day_str = '';
    idx=1;
    jdx=1;
    try:
        day_list1 = [];
        for idx in range(1,20,1):
            a1=driver.find_elements_by_xpath('//*[@id="detail-data-table"]/tbody/tr[1]/td['+str(idx)+']/div')
            for a2 in a1:
                day_str = a2.get_attribute('innerHTML');
                #print(idx, day_str)
                day_list1.append(day_str)
    except all:
        print(day_list1)
    didx=0;
    for jdx in range(1,100,1):
            b1=driver.find_elements_by_xpath('//*[@id="detail-data-table"]/tbody/tr[2]/td['+str(jdx)+']') 
            c1=driver.find_elements_by_xpath('//*[@id="detail-data-table"]/tbody/tr[6]/td['+str(jdx)+']')
            d1=driver.find_elements_by_xpath('//*[@id="detail-data-table"]/tbody/tr[7]/td['+str(jdx)+']')
            e1=driver.find_elements_by_xpath('//*[@id="detail-data-table"]/tbody/tr[8]/td['+str(jdx)+']/div')
            for b2 in b1:
                if (b2.get_attribute('class')):
                    time_str = b2.get_attribute('innerHTML');
                    time_str1=time_str.split('<small>')[0];
                    time_str2=time_str.split('<small>')[1].split('</small>')[0];
                    time_list.append(time_str1+time_str2)
                    day_list.append(day_list1[didx]);
                    didx=didx+1;
                else:
                    time_str = b2.get_attribute('innerHTML');
                    time_str1=time_str.split('<small>')[0];
                    time_str2=time_str.split('<small>')[1].split('</small>')[0];
                    time_list.append(time_str1+time_str2)
                    day_list.append(day_list1[didx]);
            for c2 in c1:
                wind_str = c2.get_attribute('innerHTML');
                wind_list.append(wind_str)
            for d2 in d1:
                gust_str = d2.get_attribute('innerHTML');
                gust_list.append(gust_str)
                lat_list.append(lat)
                long_list.append(longit)
            for e2 in e1:
                wdir_str = e2.get_attribute('style').split('(')[1].split(')')[0];
                
                wdir_list.append(wdir_str)
                        
    wind_data1 = pd.DataFrame({'latitude':lat_list,
                               'longitude':long_list,
                               'day':day_list,
                               'time (GMT)':time_list,
                               'wind speed':wind_list,
                               'wind gusts':gust_list,
                               'wind from':wdir_list})
    frames = [wind_data, wind_data1]
    result = pd.concat(frames)

    wind_data=result    
```
